```python
from photutils.detection import DAOStarFinder
from astropy.wcs import WCS
from astroquery.gaia import Gaia
from astropy.coordinates import SkyCoord
import astropy.units as u
# Function to query Gaia archive with RA and Dec
#mask = np.zeros (section. shape, dtype = bool)
#mask [1100:1400,2150:2450] = True
def query_gaia(ra, dec, radius=5.0):
    coord = SkyCoord(ra=ra*u.deg, dec=dec*u.deg, frame='icrs')
    
    # Perform cone search in Gaia archive
    job = Gaia.cone_search(coord, radius=radius*u.arcsec)
    
    # Fetch results
    result = job.get_results()
    
    return result

daofind = DAOStarFinder(fwhm = 10.0, threshold=std)#threshold = 5.0*std)#fwhm = full widht half maximum
#thresholdu bulmak strandart szpmayi 5 ile carptin. Burdaki  arpm belirlediimiz th nin hassasln ayarlyor daha yksek th demek daha ayrntl arama demek

sources = daofind(fullImage-median) #mask = mask
for col in sources.colnames:
  if col not in ("id","npix"):
    sources[col].info.format = "%.2f"
wcs = WCS(header)

# Extracting x, y coordinates of detected stars
x_coords = sources['xcentroid']
y_coords = sources['ycentroid']

# Converting pixel coordinates to celestial coordinates (RA, Dec)
ra, dec = wcs.all_pix2world(x_coords, y_coords, 0)

# Displaying the celestial coordinates

sources.pprint(max_width = 76)

indices_to_remove = []
distances = []
phot_mean_mags =[]
for i, (ra_val, dec_val) in enumerate(zip(ra, dec)):
    gaia_result = query_gaia(ra_val, dec_val)
    
    print(f"Star {i + 1}: RA = {ra_val}, Dec = {dec_val}")
    if len(gaia_result) > 0:
        # Filter out non-stellar objects based on criteria
        stellar_candidates = gaia_result[
            (gaia_result['phot_g_mean_mag'] < 20)   # Adjust magnitude threshold # Adjust color index range for stars
            #(0.5 < gaia_result['bp_rp'] < 1.5)  # Adjust color index range for stars
            #(np.abs(gaia_result['pmra']) < 10) &  # Adjust proper motion in RA threshold
            #(np.abs(gaia_result['pmdec']) < 10)   # Adjust proper motion in Dec threshold

        ]
        
        if len(stellar_candidates ) > 0:  #and float(sources['roundness1'][i] >= 0.5):
            parallax = stellar_candidates['parallax'][0]
            distance = 1.0 / parallax  # Calculate distance           
            gaia_app_mag = stellar_candidates['phot_g_mean_mag'][0]            
            flux = stellar_candidates['phot_g_mean_flux'][0]  # Extract flux
            if distance in distances and gaia_app_mag in phot_mean_mags:
                print(f"Star {i + 1}: Already in our database")
                indices_to_remove.append(i)
            else:
                phot_mean_mags.append(gaia_app_mag)
                distances.append(distance)
                print(f"Star {i + 1}: Found in Gaia archive. (Possibly a star)")
                print(f"   Distance from our Earth: {distance} parsecs")
                print(f"   Flux: {flux:.2f}")

            # Access additional information about the star in stellar_candidates
        else:
            print(f"Star {i + 1}: Found in Gaia archive, but not identified as a star based on criteria.")
            indices_to_remove.append(i)
    else:
        print(f"Star {i + 1}: Not found in Gaia archive.")
        indices_to_remove.append(i)

# Remove the rows from sources table
sources.remove_rows(indices_to_remove)

if len(sources) > 0:
    sources.pprint(max_width=76)
else:
    print("There is no match in the GAIA Archive")


#below you can see the all stars but (apply that to section1 and section2 to see differency.)
# the negative sign indicates that the shape is elongated or distorted in a particular direction. 
# The magnitude of the value (0.16) gives an indication of the extent of this distortion. However, 
# without more information about the specific measurement system or context in which this value was obtained, 
# it's challenging to provide a more detailed interpretation.
```

```python
# Besides apertures we need to draw annulus around each of star because we need to figure out what the background noise is like.
from photutils.aperture import CircularAnnulus, CircularAperture, ApertureStats, aperture_photometry

annulus_aperture = CircularAnnulus(positions, r_in = 10, r_out = 20)
plt.imshow(fullImage, cmap="Greys", norm = LogNorm(), origin = "lower")
apertures.plot(color = "blue", lw = 1.5, alpha = 0.5) # lw for line widht.
annulus_aperture.plot(color="green", lw = 1.5, alpha = 0.5) # alpha for transparency.

plt.show()
```

```python
aperstats = ApertureStats(fullImage, annulus_aperture)
bkg_mean = aperstats.mean
aperture_area = apertures.area_overlap(fullImage)
total_bkg = bkg_mean*aperture_area

star_data = aperture_photometry(fullImage, apertures)

star_data["total_bkg"] = total_bkg

for col in star_data.colnames:
  star_data[col].info.format = '%.8g'
star_data.pprint()
```

```python
#ABmag is the one u want
zero_point = 27.65
exp_time = 2700 # header da yaziyo ulasma yollari yukarda verildi.

import math

apparent_magnitudes = []

for line in star_data:
 apparent_magnitudes.append(zero_point-(2.5*math.log10(abs(line[3]-line[4]/exp_time))))

star_data["apparent_magnitudes"] = apparent_magnitudes
star_data.pprint(max_lines = -1, max_width= -1 )

def apparent_to_absolute_magnitude(apparent_magnitude, distance):

    # Calculate absolute magnitude
    absolute_magnitude = apparent_magnitude - 5 * math.log10(distance) + 5

    return absolute_magnitude
#Due to lack of information maybe there are some defects in the calculations
#but eventually this is our fully fletched draft.
a = 0
absolute_mags = []
for  apparent_mag in apparent_magnitudes:
  distance = distances[a]
  abs_mag = apparent_to_absolute_magnitude(apparent_mag,distance)
  abs_mag = "{:.2f}".format(abs_mag)
  absolute_mags.append(abs_mag)
  print(abs_mag)
star_data["absolute_mag"] = absolute_mags
star_data.pprint(max_lines = -1, max_width= -1 )
```
